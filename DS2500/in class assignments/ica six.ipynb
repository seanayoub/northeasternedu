```python
import numpy as np

quiz_array = np.array([[89, 100, 78], 
                      [75, 90, 90], 
                      [93, 85, 65],
                      [92, 92, 76]])
index = ("Quiz 1", "Quiz 2", "Quiz 3", "Quiz 4")
columns = ("Student B", "Student A", "Student C")

df_grade = pd.DataFrame(quiz_array, index, columns)
df_grade
```

```python
df_grade_subset = df_grade.iloc[1:, ::2]
df_grade_subset
```

```python
# student mean scores
df_grade_subset.mean()
```

```python
# mean score by quiz
df_grade_subset.mean(axis = 1)
```

```python
# 1
s_studentd = pd.Series({"Quiz 1": 60, "Quiz 2": 70, "Quiz 3": 80, "Quiz 4": 90})
df_grade.loc[:, "Student D"] = s_studentd
df_grade = df_grade.drop("Student D")
df_grade
```

```python
# 2
df_grade.loc["Quiz 5", :] = 1, 2, 3, 4
df_grade
```

```python
# 3
del df_grade["Student C"]
df_grade
```
