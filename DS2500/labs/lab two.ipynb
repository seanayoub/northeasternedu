# Lab 2: Triple-or-Nothing Analysis

# Loading Data




```python
import pandas as pd

df_per_game = pd.read_csv('trip_nothing_per_game_sp23.csv')
df_per_round = pd.read_csv('trip_nothing_per_round_sp23.csv')
```

## Part A:
Were shorter games associated with fewer coins earned by both players?


```python
df_per_game.groupby("round_idx").mean()
```

The data shows that shorter games **are** associated by fewer coins earned by both players.

## Part B:
Which method "won" triple-or-nothing?  You may need to make assumptions to define precisely what "winning" is.  Feel free to do so, but clearly indicate them to the grader. 

```python
import matplotlib.pyplot as plt
plt.figure()

# extract data and plot
names = df_per_round.loc[:, "name0"]
total = df_per_round.loc[:, "coin_total0"]
plt.bar(names, total, color = "r", alpha = 0.6)

# format graph
plt.xticks(rotation = "vertical")
plt.xlabel("Method Name")
plt.ylabel("Total Coins Won")
plt.title("Winning Method")
```

According to the graph, **"Misha:)"** was the winning method. The graph shows each method represented in the dataset to the total amount of coins won by each, comparing all methods after the triple-or-nothing round.
